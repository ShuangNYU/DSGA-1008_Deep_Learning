````python
cuda = torch.cuda.is_available()
device = torch.device('cuda' if cuda else 'cpu')
````

````python
class CAE(nn.Module):
    def __init__(self):
        super(CAE, self).__init__()
        # raise NotImplementedError
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(16),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d(kernel_size=2)
        )
        # self.decoder = nn.Sequential(
        #     nn.Conv2d(in_channels=32, out_channels=16, kernel_size=3, padding=1),
        #     nn.UpsamplingBilinear2d(scale_factor=2),
        #     nn.ReLU(),
        #     nn.BatchNorm2d(16),
        #     nn.Conv2d(in_channels=16, out_channels=8, kernel_size=3, padding=1),
        #     nn.UpsamplingBilinear2d(scale_factor=2),
        #     nn.ReLU(),
        #     nn.BatchNorm2d(8),
        #     nn.Conv2d(in_channels=8, out_channels=1, kernel_size=3, padding=1),
        #     nn.UpsamplingBilinear2d(scale_factor=1)
        # )
        self.decoder = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(16),
            nn.ConvTranspose2d(in_channels=16, out_channels=8, kernel_size=3, padding=1, stride=2, output_padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(8),
            nn.ConvTranspose2d(in_channels=8, out_channels=1, kernel_size=3, padding=1, stride=2, output_padding=1),
        )

    def forward(self, y):
        """Forward function of CAE class.

        Args:
          y: a batch of input images.
        
        Returns:
          yhat: the batch of reconstructed images.
        """
        # raise NotImplementedError
        enc_y = self.encoder(y)
        out_y = self.decoder(enc_y)
        return out_y
````

````python
# This is part of question 4.
def add_noise(images):
    # raise NotImplementedError
    std_Gaussian = torch.distributions.normal.Normal(torch.tensor([0.0]), torch.tensor([1.0]))
    uniform = torch.distributions.uniform.Uniform(torch.tensor([0.0]), torch.tensor([1.0]))
    images_noisy = images.clone()
    for i, image in enumerate(images):
      noise = std_Gaussian.sample(image.shape).squeeze(-1)
      coef = uniform.sample()
      images_noisy[i] = torch.clamp(image + noise * coef, min = -1.0, max = 1.0)

    return images_noisy
````

````python
def train(autoencoder, train_loader, epoch, num_epoch, learning_rate,
          log_interval, denoising=False):
    # Using sum reduction simply bc otherwise, the scale of the loss is too small;
    # given that we always have 28*28 images, using sum/mean reduction wouldn't
    # make a difference.
    criteria = nn.MSELoss(reduction='sum')
    optimizer = torch.optim.Adam(
        autoencoder.parameters(), lr=learning_rate)

    autoencoder.train()
    for i, data in enumerate(train_loader):
        images, _ = data
        if denoising:
            input_images = add_noise(images)
        else:
            input_images = images
        if cuda:
            input_images = input_images.to(device)
            images = images.to(device)
        outputs = autoencoder(input_images)
        loss = criteria(outputs, images)
        # total_loss += loss
        autoencoder.zero_grad()
        loss.backward()
        optimizer.step()

    # avg_loss = total_loss / (len(valid_loader) * batch_size)

        
        # raise NotImplementedError
        if (i + 1) % log_interval == 0:
            print('Epoch [{}/{}] - Iter[{}/{}], MSE loss:{:.6f}'.format(
                epoch + 1, num_epoch, i + 1,
                len(train_loader.dataset) // batch_size, loss.item() / batch_size
            ))


def validate(autoencoder, valid_loader, best_val_loss, denoising=False):
    total_loss = 0
    criteria = nn.MSELoss(reduction='sum')
    autoencoder.eval()
    for i, data in enumerate(valid_loader):
        images, _ = data
        if denoising:
            input_images = add_noise(images)
        else:
            input_images = images
        if cuda:
            input_images = input_images.to(device)
            images = images.to(device)
        outputs = autoencoder(input_images)
        loss = criteria(outputs, images)
        total_loss += loss
    avg_loss = total_loss / (len(valid_loader) * batch_size)

    print('\nLoss per batch on validation set: {:.6f}'.format(avg_loss))

    if avg_loss < best_val_loss:
        if denoising:
            save_path = '/content/drive/My Drive/1008_DL_Homework3/CAE_noisy/best_model.pt'
        else:
            save_path = '/content/drive/My Drive/1008_DL_Homework3/CAE_original/best_model.pt'
        best_val_loss = avg_loss
        torch.save({
                  'model': autoencoder.state_dict(),
                  # 'optimizer': optimizer.state_dict(),
                  # 'epoch': t,
                  'best_val_loss': best_val_loss
              }, save_path)
        # raise NotImplementedError
        print('Saved best model in the checkpoint directory\n')
    
    return best_val_loss

def evaluate(autoencoder, test_loader, denoising=False):
    total_loss = 0
    criteria = nn.MSELoss(reduction='sum')
    autoencoder.eval()
    for i, data in enumerate(test_loader):
        images, _ = data
        if denoising:
            input_images = add_noise(images)
        else:
            input_images = images
        if cuda:
            input_images = input_images.to(device)
            images = images.to(device)
        outputs = autoencoder(input_images)
        loss = criteria(outputs, images)
        total_loss += loss
    avg_loss = total_loss / (len(test_loader) * batch_size)

    print('\nMSE loss per batch on test set: {:.6f}'.format(avg_loss))
````

````python
def visualize_one_model(model, image0, image1, gamma1, images):
    # raise NotImplementedError
    model_children = list(model.children())
    encoder, decoder = model_children[0], model_children[1]
    z0 = encoder(image0.unsqueeze(0))
    z1 = encoder(image1.unsqueeze(0))
    z = gamma1*z0+(1-gamma1)*z1
    image = torch.clamp(decoder(z), min = -1.0, max = 1.0)
    # imshow(torchvision.utils.make_grid(image.data))
    images.append(image)
    if len(images) == 10:
      images = torch.stack((images[0], images[1], images[2], images[3], images[4],
             images[5], images[6], images[7], images[8], images[9])).squeeze()
      images = images.unsqueeze(1)
      imshow(torchvision.utils.make_grid(images.data, nrow = 10))
````

````python
images = []
for k in np.arange(0, 1.01, 0.1):
  print('k =', k)
  image = visualize_one_model(autoencoder, image0, image1, k, images)
````

From left to right, the images correspond to k gradually increased. That consists with our expectation!
